### The dot porduct
Matrix multiplication is also called as dot product
1. tf.matmul()
2. tf.tensordot()
3. @

```python
# perform dot product on X and Y, this X or Y to be transposed
X=tf.constant([[1,2],[3,4],[5,6]])
Y=tf.constant([[7,8],[9,10],[11,12]])
print(X,Y)
tf.tensordot(tf.transpose(X),Y,axes=1)
```

```python
# Perform matrix multiplication X and Y(transposed)
tf.matmul(X,tf.transpose(Y))
```

```python
# Perform matrix multiplication X and Y(reshaped)
tf.matmul(X,tf.reshape(Y,shape=(2,3)))
```

```python
# Check values of Y and transposed Y
print("Normal Y: \n",Y)
print("\nY reshaped: \n",tf.reshape(Y,shape=(2,3)))
print("\nY Transposed: \n",tf.transpose(Y))
```

### Aggregating tensors
Aggregating tensors = condensing them from multiple values down to a small amount of values

```python
# Getting the absolute values
D = tf.constant([-7,-10])
print(D)
tf.abs(D)
```

#### forms of aggregation:
1. Get the minimum of tensor
2. Get the maximum of tensor
3. Get the mean of the tensor
4. Get the sum of the tensor

```python
test = tf.constant(np.random.randint(0,100,size=50))
print(test)
print("Minimum: ",tf.reduce_min(test).numpy())
print("Maximum: ",tf.reduce_max(test).numpy())
print("mean: ",tf.reduce_mean(test).numpy())
print("Sum of all elements: ",tf.reduce_sum(test).numpy())
print("Varience: ",tf.math.reduce_variance(tf.cast(test,dtype=tf.float32)).numpy())
print("Standard deviation: ",tf.math.reduce_std(tf.cast(test,dtype=tf.float32)).numpy())
```

```python
#another method to find varience and std:
import tensorflow_probability as tfp
tfp.stats.variance(test)
```

```python
# FInd the positional maximum and minimum of the tensor
print("Positional maximun: ",tf.argmax(test).numpy(),":",test[tf.argmax(test).numpy()].numpy())
print("Positional minimun: ",tf.argmin(test).numpy(),":",test[tf.argmin(test).numpy()].numpy())
```
